```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, recall_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df = pd.read_csv(r"C:\Users\drago\Downloads\archive (4)\emails.csv")

df
```

```python
df.head()
```

```python
df = df[['text', 'spam']].fillna('')

df
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.2, random_state=42)

```

```python
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

```

```python
nb_classifier1 = MultinomialNB()
nb_classifier2 = MultinomialNB()
nb_classifier3 = MultinomialNB()

ensemble = VotingClassifier(estimators=[
    ('nb1', nb_classifier1),
    ('nb2', nb_classifier2),
    ('nb3', nb_classifier3)
], voting='hard')  
```

```python
ensemble.fit(X_train_vectorized, y_train)

y_pred = ensemble.predict(X_test_vectorized)

```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100.:2f}')
```
