```python
import pandas as pd

transactions = [
    ['bread', 'milk', 'vegetables'],
    ['bread', 'milk', 'cheese'],
    ['bread', 'butter'],
    ['milk', 'cheese'],
    ['bread', 'milk', 'cheese', 'butter'],
    ['bread', 'milk'],
    ['bread', 'milk', 'cheese'],
    ['bread', 'eggs'],
    ['bread', 'eggs', 'butter'],
    ['eggs', 'milk', 'cheese'],
    ['bread', 'butter', 'cheese'],
    ['bread', 'milk', 'butter'],
    ['bread', 'eggs', 'milk'],
    ['bread', 'milk', 'cheese', 'butter'],
    ['bread', 'milk', 'eggs']
]

```

```python
unique_items = list(set(item for transaction in transactions for item in transaction))

item_to_index = {item: i for i, item in enumerate(unique_items)}

matrix = [[0] * len(unique_items) for _ in range(len(transactions))]

for i, transaction in enumerate(transactions):
    for item in transaction:
        matrix[i][item_to_index[item]] = 1

```
