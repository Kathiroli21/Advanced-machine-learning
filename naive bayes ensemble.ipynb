```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Load your data
df = pd.read_csv(r"C:\Users\drago\Downloads\archive (4)\emails.csv")

# Assuming the 'text' column contains the email content and 'spam' column is the label (1 for spam, 0 for not spam)

# Drop unnecessary unnamed columns with NaN values
df = df[['text', 'spam']]

# **1. Consistent Type Conversion in Data Cleaning:**
# Ensure conversion to string only if needed
df['text'] = df['text'].apply(lambda x: x if isinstance(x, str) else '')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.2, random_state=42)

# Create a CountVectorizer to convert text data into numerical features
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Instantiate Gaussian Naive Bayes classifier
naive_bayes = GaussianNB()

# Ensemble using VotingClassifier
ensemble_voting = VotingClassifier(estimators=[('nb', naive_bayes)], voting='hard')
ensemble_voting.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using BaggingClassifier
ensemble_bagging = BaggingClassifier(base_estimator=naive_bayes, n_estimators=10, random_state=42)
ensemble_bagging.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using AdaBoostClassifier
ensemble_adaboost = AdaBoostClassifier(base_estimator=naive_bayes, n_estimators=50, random_state=42)
ensemble_adaboost.fit(X_train_vectorized.toarray(), y_train)

# Predictions on the test set
y_pred_voting = ensemble_voting.predict(X_test_vectorized.toarray())
y_pred_bagging = ensemble_bagging.predict(X_test_vectorized.toarray())
y_pred_adaboost = ensemble_adaboost.predict(X_test_vectorized.toarray())

# Evaluate accuracy
accuracy_voting = accuracy_score(y_test, y_pred_voting)
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)

print("Voting Classifier Accuracy:", accuracy_voting)
print("Bagging Classifier Accuracy:", accuracy_bagging)
print("AdaBoost Classifier Accuracy:", accuracy_adaboost)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Load your data
df=pd.read_csv(r"C:\Users\drago\Downloads\archive (4)\emails.csv")

# Assuming the 'text' column contains the email content and 'spam' column is the label (1 for spam, 0 for not spam)

# Drop unnecessary unnamed columns with NaN values
df = df[['text', 'spam']]

# Replace NaN values in the 'text' column with empty strings
df['text'] = df['text'].fillna('')

# Convert all values in the 'text' column to strings
df['text'] = df['text'].astype(str)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.2, random_state=42)

# Create a CountVectorizer to convert text data into numerical features
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Instantiate Gaussian Naive Bayes classifier
naive_bayes = GaussianNB()

# Ensemble using VotingClassifier
ensemble_voting = VotingClassifier(estimators=[('nb', naive_bayes)], voting='hard')
ensemble_voting.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using BaggingClassifier
ensemble_bagging = BaggingClassifier(base_estimator=naive_bayes, n_estimators=10, random_state=42)
ensemble_bagging.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using AdaBoostClassifier
ensemble_adaboost = AdaBoostClassifier(base_estimator=naive_bayes, n_estimators=50, random_state=42)
ensemble_adaboost.fit(X_train_vectorized.toarray(), y_train)

# Predictions on the test set
y_pred_voting = ensemble_voting.predict(X_test_vectorized.toarray())
y_pred_bagging = ensemble_bagging.predict(X_test_vectorized.toarray())
y_pred_adaboost = ensemble_adaboost.predict(X_test_vectorized.toarray())

# Evaluate accuracy
accuracy_voting = accuracy_score(y_test, y_pred_voting)
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)

print("Voting Classifier Accuracy:", accuracy_voting)
print("Bagging Classifier Accuracy:", accuracy_bagging)
print("AdaBoost Classifier Accuracy:", accuracy_adaboost)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils.multiclass import type_of_target

# Load your data
df = pd.read_csv(r"C:\Users\drago\Downloads\archive (4)\emails.csv")

# Assuming the 'text' column contains the email content and 'spam' column is the label (1 for spam, 0 for not spam)

# Drop unnecessary unnamed columns with NaN values
df = df[['text', 'spam']]

# Ensure consistent type conversion in Data Cleaning
df['text'] = df['text'].apply(lambda x: str(x) if isinstance(x, (float, int)) else '')

# Check and convert target variable to integer type if needed
if type_of_target(df['spam']) == 'binary':
    df['spam'] = df['spam'].astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.2, random_state=42)

# Create a CountVectorizer to convert text data into numerical features
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Instantiate Gaussian Naive Bayes classifier
naive_bayes = GaussianNB()

# Ensemble using VotingClassifier
ensemble_voting = VotingClassifier(estimators=[('nb', naive_bayes)], voting='hard')
ensemble_voting.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using BaggingClassifier
ensemble_bagging = BaggingClassifier(base_estimator=naive_bayes, n_estimators=10, random_state=42)
ensemble_bagging.fit(X_train_vectorized.toarray(), y_train)

# Ensemble using AdaBoostClassifier
ensemble_adaboost = AdaBoostClassifier(base_estimator=naive_bayes, n_estimators=50, random_state=42)
ensemble_adaboost.fit(X_train_vectorized.toarray(), y_train)

# Predictions on the test set
y_pred_voting = ensemble_voting.predict(X_test_vectorized.toarray())
y_pred_bagging = ensemble_bagging.predict(X_test_vectorized.toarray())
y_pred_adaboost = ensemble_adaboost.predict(X_test_vectorized.toarray())

# Evaluate accuracy
accuracy_voting = accuracy_score(y_test, y_pred_voting)
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)

print("Voting Classifier Accuracy:", accuracy_voting)
print("Bagging Classifier Accuracy:", accuracy_bagging)
print("AdaBoost Classifier Accuracy:", accuracy_adaboost)
```
